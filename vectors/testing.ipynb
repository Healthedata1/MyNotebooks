```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1,0])
v2 = np.array([0,1])
v3 = np.array([np.sqrt(2), np.sqrt(2)])
```

#### Plot of Vectors

```python

origin = np.array([0, 0])

plt.figure(figsize=(10, 10))
plt.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='b', label='v2')
plt.quiver(*origin, *v3, angles='xy', scale_units='xy', scale=1, color='g', label='v3')

plt.xlim(-2.5, 2.5)
plt.ylim(-1.5, 2)
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal')
plt.title("Visualizing Vectors in 2D")
plt.show()
```

### Magnitude  or Norm (a non-negative number that represents the vector’s size or length)
 - but the most common is the Euclidean norm

```python
np.sqrt(np.sum(v1**2))
```

- a NumPy function that computes the Euclidean norm of an array by default but can also compute other matrix and vector norms.

```python
np.linalg.norm(v1)
```

### Dot product (it measures the similarity between two vectors , and is a number given by u ⋅ v = ||u|| ||v|| cos(θ))

#### Orthogonal Vectors 
  - Another way to compute the dot product is to do an element-wise multiplication of u and v and sum the results. 

```python
np.sum(v1 * v2)
```

#### The dot product of two vectors
  -  @ can perform both vector and matrix multiplications

```python
v1 @ v3
```

Cosine Similarity  = normalizes relationship bewteen 2 vectors

```python
(v2 @ v3) / (np.linalg.norm(v2) * np.linalg.norm(v3))
```
