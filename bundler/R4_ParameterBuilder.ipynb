###  write to file

```python
def write_file(r): # write file
    #out_path = '/Users/ehaas/Documents/FHIR/Davinci-DEQM/source/examples/'  # append forward slash
    out_path = in_path
    with open(f'{out_path}/Parameter-{parameters_id}.json', 'w') as f:
        f.write(r)
```

### open json file and return as dict

```python
def open_file(f_name): # get files
    with open(f'{in_path}/{f_name}') as f:
        r = f.read()
        return(loads(r))
```

### create Parameters 'p' change the id for unique Bundles!!!

```python
p_resource = {'id': parameters_id, 'resourceType': 'Parameters'}
```

### Add resources to Parameters Resource



- get file from folder based on list


- add to parameters resource


```python
p_resource['parameter'] = []

p_name = "measure-report"
for i in files:
    p = {}
    p['name'] = p_name
    p['resource'] = open_file(i)
    p_resource['parameter'].append(p)
    p_name = "resource"
    
    
p_json = dumps(p_resource,indent =4)
print(p_json)

write_file(p_json)
```
