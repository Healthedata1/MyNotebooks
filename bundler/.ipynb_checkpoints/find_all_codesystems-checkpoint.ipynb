###  Read all FHIR resources and find all the strong bindings for elements with max = "*"

```python
from pathlib import Path
from json import load, dumps, loads
```

```python
#path = Path(r'/Users/ehaas/Downloads/definitions.json/profiles-resources.json')
path = Path(r'/Users/ehaas/Downloads/definitions.json/extension-definitions.json')
b_dict=loads(path.read_text())
sds = [sd['resource'] for sd in b_dict['entry'] if sd['resource']['resourceType']=="StructureDefinition"] 
```

```python
for sd in sds:
    #print(f"====== {sd['name']} ======") 
    my_e = [e['id'] for e in sd['snapshot']['element'] if e['max']=='*']
            #e.get('binding',{}).get("strength") in ['extensible', 'required']
            #and e['max']=='*']
    #print(my_e)
    for e in my_e:
        print(f'1. {sd["name"]}.{e}')
```
