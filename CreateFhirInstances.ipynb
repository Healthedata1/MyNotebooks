```python
from importlib import import_module
from os import listdir
from json import load, dumps, loads
from IPython import display as D
from requests import get, post, put
from IPython.display import display, Markdown, HTML
```

### Get files

```python
def open_file(in_path,f_name): # get files
    with open(f'{in_path}{f_name}',encoding='utf-8', errors='ignore') as f:
        r = f.read()
       
        return(loads(r))
```

### Get fhirmodel class and instantiate

```python
def get_model(Type, data):
    # Instantiate the class (pass arguments to the constructor, if needed)
    m = getattr(import_module(f'fhirclient.r4models.{Type.lower()}'), Type)
    return(m(data, strict=False))

```

## get all files in folder and convert to python fhir model

```python
in_path='unbundle_save/'
for file in listdir(in_path):
    if file.endswith(".json"):
        print("json file found:\t", file)
        b = open_file(in_path,file)
        fm = get_model(b['resourceType'],b)
        print(fm.resource_type, fm.id)
```
