### TEST JQ

- move array items to new key

```python
%%bash
JsonData='{"name":"foo","array": [{"name": "bar","thing": true},{"name": "baz","thing": false},{"name": "bar","thing": "foo"}],"other": {"foo": "bar"}}'
echo $JsonData | jq '.'
echo "========add new array======="
echo $JsonData | jq '.other += { "new_array": [] }'
echo "========select old array element======="
echo $JsonData | jq '.array | .[] | select(.name == "bar")' 
# echo "========delete array element======="
# echo $JsonData | jq 'del(.array[] | select(.name == "bar"))' 
# echo "========move array element======="
# echo $JsonData | jq '.other += { "new_array": [.array[] | select(.name == "bar")] }'
echo "========move and delete array element======="
echo $JsonData | jq '.other += { "new_array": [.array[] | select(.name == "bar")] } | del(.array[] | select(.name == "bar"))'
```
