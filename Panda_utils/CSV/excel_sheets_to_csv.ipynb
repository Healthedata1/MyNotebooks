### Read an Excel Workbook and Save Each Sheet as new csv

- Read Excel from file
- for each sheet:
    - change headers
    - save as CSV

```python
from pandas import *
from pathlib import Path
from IPython.core.display import HTML
```

```python
filename_map = { 
"omh-header-fhir-observation-map": "omh-header-fhir-observation-mapping",
"ieee-header-fhir-observation-ma": "ieee-header-fhir-observation-mapping",
"ieee-header-fhir-provenance-map": "ieee-header-fhir-provenance-mapping",
"omh-header-fhir-provenance-mapp": "omh-header-fhir-provenance-mapping",
"omh-heartrate-fhir-observation-": "omh-heartrate-fhir-observation-mapping",
"omh-resprate-fhir-observation-m": "omh-resprate-fhir-observation-mapping",
"omh-bodytemp-fhir-observation-m": "omh-bodytemp-fhir-observation-mapping",
"omh-bloodpressure-fhir-observat": "omh-bloodpressure-fhir-observation-mapping",
"omh-bodyweight-fhir-observation": "omh-bodyweight-fhir-observation-mapping",
"omh-sleeptime-fhir-observation-": "omh-sleeptime-fhir-observation-mapping",
"omh-stepcount-fhir-observation-": "omh-stepcount-fhir-observation-mapping",
"omh-bloodglucose-fhir-observati": "omh-bloodglucose-fhir-observation-mapping",
"omh-o2sat-fhir-observation-mapp": "omh-o2sat-fhir-observation-mapping",
}
```

### Get all sheets from Excel file

```python
my_path = '/Users/ehaas/Documents/FHIR/OMHtoFHIR/mapping/mapping tables'
my_file = 'json-schema-fhir-mappings.xlsx'
excel_wb_file = Path()/my_path/my_file
df = read_excel(excel_wb_file, sheet_name=None)
for i in df.keys(): # = sheet names
    
    print(filename_map[i])
    display(HTML(df[i].head().to_html()))
    print('='*80)
    new_headers = df[i].iloc[0]
    print(new_headers)
    df2 = df[i][2:]
    df2.columns = new_headers
    display(HTML(df2.to_html()))
    new_file_name = f"{filename_map[i]}.csv"
    new_path = r'/Users/ehaas/Documents/Python/MyNotebooks/Panda_utils/CSV/my_csv'
    new_file = Path(new_path)/new_file_name 
    df2.to_csv(new_file, index=False)
    print(f'printing {new_file_name} to {new_file}')
```
