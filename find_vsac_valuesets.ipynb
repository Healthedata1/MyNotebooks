### Find all valueset from VSAC

1. canonical will start with "http://cts.nlm.nih.gov/"
2. cycle through all the StructureDefinition files in the package file here /Users/ehaas/.fhir/packages/hl7.fhir.us.core#6.1.0/package/
  for example /Users/ehaas/.fhir/packages/hl7.fhir.us.core#6.1.0/package/StructureDefinition-us-core-respiratory-rate.json
3. use Python pathlib
4 produce a simple list
5. remove duplicates.

```python
from pathlib import Path
import json
import re
import sys

# ---------------------------------------------------
# CONFIGURATION
# ---------------------------------------------------
# Update this path if your US Core package is elsewhere
PACKAGE_DIR = Path("/Users/ehaas/.fhir/packages/hl7.fhir.us.core#6.1.0/package")

# Output file (will be created in the same folder as this script)
# OUTPUT_FILE = Path(__file__).with_name("vsac_valuesets.txt")

# ---------------------------------------------------
# MAIN LOGIC
# ---------------------------------------------------
def main():
    # Find all StructureDefinition JSON files
    sd_files = list(PACKAGE_DIR.glob("StructureDefinition-*.json"))
    print(f"Found {len(sd_files)} StructureDefinition files.\n")

    # Regex to match VSAC ValueSet URLs
    # Examples:
    #   http://cts.nlm.nih.gov/fhir/ValueSet/2.16.840.1.113883.3.464.1003.110.12.1053
    #   https://cts.nlm.nih.gov/fhir/ValueSet/...
    pattern = re.compile(
        r"https?://cts[.-]nlm[.]nih[.]gov/fhir/ValueSet/[A-Za-z0-9.-]+"
    )
    pattern_count = 0
    sub = "cts.nlm.nih.gov/fhir/ValueSet/"
    # print(sub)
    sub_count = 0

    vsac_urls = set()

    # Scan each file
    for file_path in sd_files:
        # print(file_path.name)
        try:
            data = json.loads(file_path.read_text(encoding="utf-8"))
            text = json.dumps(data)  # Convert to string for regex search
            matches = pattern.findall(text)
            # print(matches)
            pattern_count += len(matches)
            sub_count += text.count(sub)
            # print(sub_count)

            if matches:
                count_before = len(vsac_urls)
                vsac_urls.update(matches)
                new_count = len(vsac_urls) - count_before
                print(f"{file_path.name}: {len(matches)} match(es) → {new_count} new unique")

        except json.JSONDecodeError as e:
            print(f"JSON error in {file_path.name}: {e}")
        except Exception as e:
            print(f"Error reading {file_path.name}: {e}")

    # ---------------------------------------------------
    # OUTPUT
    # ---------------------------------------------------

    print("\n" + "="*60)
    print(f" TOTA VALUESETS: pattern_count={pattern_count}, sub_count={sub_count} ")
    print("="*60)

    unique_urls = sorted(vsac_urls)

    print("\n" + "="*60)
    print(f" TOTAL UNIQUE VSAC VALUESETS: {len(unique_urls)} ")
    print("="*60)

    for i, url in enumerate(unique_urls, 1):
        print(f"{i:3}. {url}")

    # # Save to file
    # try:
    #     OUTPUT_FILE.write_text("\n".join(unique_urls) + "\n", encoding="utf-8")
    #     print(f"\nSaved to: {OUTPUT_FILE.resolve()}")
    # except Exception as e:
    #     print(f"Failed to write output file: {e}")

if __name__ == "__main__":
    main()
```
