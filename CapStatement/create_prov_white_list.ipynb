## create provenance US Core White List from cap statement spreadsheet

- get excel file
- convert resources sheet to pandas
- iterate through rows
  - if shall_revinclude = Provenance:target
   - add to list
-write list to file



```python
#imports

from pathlib import Path
from pandas import *
```

```python
# globals

ig_source_path =''
#ig_source_path = "/Users/ehaas/Documents/FHIR/US-Core-R4/source/" # mac
#ig_source_path  = '//ERICS-AIR-2/ehaas/Documents/FHIR/US-Core-R4/source/' # pc

in_path = Path(r"\\ERICS-AIR-2\ehaas\Documents\FHIR\US-Core-R4\source\source_spreadsheets\uscore-server.xlsx") #/ in_file
out_path = Path.cwd() / ig_source_path  / 'pages' /'_includes' / 'prov-white-list.md'
in_path, out_path
```

```python
xls = ExcelFile(in_path)
df = read_excel(xls,'resources',na_filter = False)

df
```

```python
type_list = df[df['shall_revinclude'] == "Provenance:target"]["type"]
type_list
```

```python
xls = ExcelFile(in_path)
df2 = read_excel(xls,'profiles',na_filter = False)

df2
```

```python
prov_white_list = df2[df2.Type.isin(type_list)]['Name'].sort_values()
```

```python
print(".....writing to file ...........")
#out_path.write_text('- The US Core Provenance resource **SHALL** be supported for these US Core resources:\n')
out_path.write_text('\n')  # delete file
with out_path.open('a', encoding='utf-8', newline='\n') as fp:
    for i in prov_white_list:
        fp.write(f'   - {i}\n')
    fp.write(f'   - The Vital Signs Profile from the FHIR Specification')
#http://localhost:8888/lab/tree/CapStatement/pages/_includes/prov-white-list.md
```
