```python
def get_snippet(s):
    delimiters = [',',']','}','\n']
    for i in range(800,870):
        #print(i,s[i])
        if s[i] in delimiters:
            break

    return f'''{s[:i+1]}'''
```

```python
def get_md_string_del(r):
    return f'''

**Request**

~~~
{r.request.method} {r.url}

User-Agent: {r.request.headers["User-Agent"]}
Accept: {r.request.headers["Accept"]}
Content-Type: {r.request.headers["Content-Type"]}
[other headers]
~~~

**Response**

~~~
Response Code: {r.status_code} ({r.reason})
[other headers]
~~~
'''
```

```python
def get_md_string(r):
    r_json = dumps(r.json(),indent=3)
    #print(r_json)

    return f'''

**Request**

~~~
{r.request.method} {r.url}

User-Agent: {r.request.headers["User-Agent"]}
Accept: {r.request.headers["Accept"]}
Content-Type: {r.request.headers["Content-Type"]}
[other headers]
~~~

**Response**

~~~
Response Code: {r.status_code} ({r.reason})
Content-Type: {r.headers["Content-Type"]}
[other headers]

    
{get_snippet(r_json)}
        ...[snipped for brevity]....
      
~~~
'''  
```

```python
def fetch(action, id, params, r_type, data):
    fhir_test_server = 'http://sqlonfhir-stu3.azurewebsites.net/fhir'
    
    headers = {
    'Accept':'application/fhir+json',
    'Content-Type':'application/fhir+json'
    }
    f_url = f'{fhir_test_server}/{r_type}/{id}' if id==id else f'{fhir_test_server}/{r_type}'
    
    if r_type == None:
        f_url = f'{fhir_test_server}'
    
    r = requests.request(action, f_url, params = params, headers = headers,data=data)
    print(r.status_code)
    print(r.reason)
    #print(f'{r.request.method} {r.url}')
    #for k,v in r.headers.items():
    #   print(f'{k} = {v}')
    #for k,v in r.json().items():
    #print(f'{k} = {v}')
    return r
```

```python
for e in examples[17:]:  # slice on the latest additions to spreadsheet
    if e['r_type'] == 'q':
        r_type = 'Questionnaire'
    elif e['r_type'] == 'qr':
        r_type = 'QuestionnaireResponse'
    elif e['r_type'] != e['r_type']:
        r_type = None
    
    data={}
    if e['data'] == e['data']:
        data = get_data(e['data'])
    #print(data)
       
    params={}
    for i in range(1,4):
        #print(i)
        if e[f'parameter_{i}'] == e[f'parameter_{i}']:  # not nan
            params[e[f'parameter_{i}']] = e[f'value_{i}']
    print(f'params={params}')
    #continue
    r = fetch(e['action'],id=e['id'],params=params,r_type=r_type, data=data)
    
    # print(r.text)
    

    #print(example_string)
    md_str = get_md_string(r) if e['action'] != "DELETE" else get_md_string_del(r) 
    # display(D.Markdown(f'<h4>Example</h4>{md_str}'))
    
    write_file(name = f'example-{e["name"]}',data = md_str)
  
```
