```python
import json
import jcs
```

```python
my_json_object1 = '{"a":1,"c": 2}'
my_json_object2 = '{"z":"foo","x":"bar"}'
print(my_json_object1,my_json_object2)
my_objects = []
my_objects.append(json.loads(my_json_object1))
my_objects.append(json.loads(my_json_object2))

"""
print(json.dumps(my_object))
print(json.dumps(my_object, sort_keys=True,separators=(',', ':')))
print(json.dumps(my_object, sort_keys=True,separators=(',', ':')).encode().hex())
"""
my_objects
```

```python
for i,my_object in enumerate(my_objects):
  # print(my_object)
  data = jcs.canonicalize(my_object, utf8=True)
  concat_data = data + data
  print(data, data.hex(), concat_data)
```

```python
data = jcs.canonicalize(json.loads('{"tag":4}'))
data2 = json.dumps(json.loads('{"tag":4}'), sort_keys=True,separators=(',', ':'))
                   
data.hex() == data2.encode().hex()
```
