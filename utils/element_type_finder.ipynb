```python
def get_targetProfile(r_type):
    for sd_base in R4_path.glob(f'StructureDefinition*-{r_type}.json'):
        try:
            r_base = loads(sd_base.read_text())
        except Exception as e:
            print("Exception: {}".format(type(e).__name__))
            print("Exception message: {}".format(e))
        # print(f"======{r_base['type']}======")
        for element in r_base['differential']['element']:
            try:
                for  my_type in element['type']:
                    if my_type['code'] == 'Identifier':
                        print(f'''- type: element  
  expression: {element['path']}''')
                    if my_type['code'] == 'Reference':
                        print(f'''- type: element  
  expression: {element['path']}.identifier''')
            except KeyError:
                # print(f"no type in {element['path']}")
                pass
        
            
for entity in entities:
    get_targetProfile(entity)
```
