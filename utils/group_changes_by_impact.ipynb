```python
# for each line in data, find  adn insert the change impact from the data_df
impact_list = []
for line in data.strip().split('\n'):
    start = line.find('[FHIR-') + 1
    end = line.find(']', start)
    issue_key = line[start:end].split('/')[-1]
    print(f'Processing issue key: {issue_key}')

    # find the change impact from the data_df
    impact = data_df.loc[data_df['Key'] == issue_key, 'Change Impact'].values
    change_category = data_df.loc[data_df['Key'] == issue_key, 'Change Category'].values
    if len(impact) > 0:
        if impact[0] != impact[0]:  # Handles NaN, None, or other missing values
            impact[0] = "Non-substantive"
        if change_category[0] != change_category[0]:  # Handles NaN, None, or other missing values
            change_category[0] = "Correction"
        print(f'Impact for issue key {issue_key}: {impact[0]}')
        print(f'Change catefory for issue key {issue_key}: {change_category[0]}')
        # print(f'line: {line.replace("** (", f"** {impact[0]} (")}')
        if change_category[0] not in line:
            print(f'change category not in line')
            if "Applied" not in line:
                new_line = line.replace('1.', f"1. **Applied:** {impact[0]} ({change_category[0]}) ",1)
            else:
                new_line = line.replace('**:',':**',1).replace(":**", f":** {impact[0]} ({change_category[0]}) ")
        else:
            new_line = line.replace("** (", f"** {impact[0]} (")
    else:
        # impact_list.append((line, 'Not Found'))
        print(f'Impact not found for issue key: {issue_key}')
        new_line = line
    impact_list.append(new_line)
# print the updated lines
for updated_line in impact_list:
    print(updated_line)
```


### Function to extract FHIR number from a line for sorting


```python
def extract_fhir_number(line):
    match = re.search(r'FHIR-(\d+)', line)
    return int(match.group(1)) if match else 0
```


### Step 1: Parse the input data, preserving subitems


```python
# lines = data.split('\n')
lines= impact_list
items = []
current_item = None
# current_subitems = []

for line in lines:
    line = line.strip()
    if not line:
        continue
    # Check if the line is a top-level item (starts with a number and a dot)
    if re.match(r'^\d+\.\s', line):
        # Save the previous item and its subitems
        if current_item:
            items.append(current_item)
        current_item = line
        # current_subitems = []
    # elif line.startswith('  - ') or line.startswith(' - '):
    #     # Add subitem to the current item's subitems
    #     current_subitems.append(line)
    else:
        # If the line is not a subitem but not a new item, treat it as part of the current item
        if current_item:
            current_item += '\n' + line

# Save the last item
if current_item:
    items.append(current_item)#, current_subitems))


items  # Display the first two items to verify parsing
```


### Step 2: Group items by Change Impact


```python
grouped_items = defaultdict(list)
for item in items:
    if 'Non-substantive' in item:
        grouped_items['Non-substantive'].append((item))
    elif 'Compatible, substantive' in item:
        grouped_items['Compatible, substantive'].append((item))
    elif 'Non-compatible' in item:
        grouped_items['Non-compatible'].append((item))
    else:
        grouped_items['Unknown'].append((item))
```


### Step 3: Sort each group by FHIR number


```python
for category in grouped_items:
    grouped_items[category].sort(key=lambda x: extract_fhir_number(x[0]))
```


### Step 4: Generate Markdown output with sequential numbering across groups


```python

output = "**Tracker Status**: **Change Impact** **(Change Category)** **Change Summary** **Jira Issue [Link](#)** **Updated Content [Link](#)**\n\n"
categories = ['Non-compatible','Compatible, substantive', 'Non-substantive',  'Unknown']
item_number = 1  # Initialize sequential numbering

for category in categories:
    if category in grouped_items:
        output += f"#### {category}\n"
        for item in grouped_items[category]:
            # print(item)
            # Remove the original number (e.g., '1. ', '2. ') and replace with sequential number
            item_text = re.sub(r'^\d+\.\s', f'{item_number}. ', item)
            output += f"{item_text.replace('- ','    - ').replace('**Resolved     - change required:**','**Resolved - change required:**')}\n"
            # for subitem in subitems:
            #     # print(subitem)
            #     output += f"{subitem}\n"
            # output += ""
            item_number += 1  # Increment the sequential number
```
