# List US Core Invariants

- Fetch SD file from IG/input
- Sort through all the differentials
- Find all invariants
- sort and list in output

```python
from pathlib import Path
in_path = Path('/Users/ehaas/Documents/FHIR/US-Core/input/resources')
constraints = []
for i, f in enumerate(in_path.glob("Struct*.*")):
    # print(i, f)
    sd = loads(f.read_text())
    for e in sd['differential']['element']:
          try:
             
             e_constraints = e['constraint']
             for constraint in e_constraints:
              # print(constraint)
              constraints.append(constraint)
          except:
             pass
          
# constraints = sorted(constraints, key=lambda x:x['key'].split('-')[-1])
dset = set(dumps(d) for d in constraints)
result = [loads(d) for d in dset]
result = sorted(result, key=lambda x:int(x['key'].split('-')[-1]))
result
for r in result:
   print(f"{r['key']}: {r['human']}")
```
