

## Find new profiles for each version 

- Fetch package for IG latest version (dev)
- list all SDs
- repeat for older IG version
- compare

*Note need a successful build to generate since based on ig output local file
alternatively use package.files to generate*

### import python modules

```python
from pathlib import Path
from json import loads, dumps
from fhir.resources.structuredefinition import StructureDefinition
from fhir.resources.elementdefinition import ElementDefinition
from pprint import pprint
```

### get package

```python
import warnings  #suppress pandas futures warnings
warnings.filterwarnings("ignore")

new_ig_profiles = []
old_ig_profiles = []
old_version = '7.0.0'
new_pkg = Path(r'/Users/ehaas/.fhir/packages/hl7.fhir.us.core#dev/package') #latest version
old_pkg = Path(f'/Users/ehaas/.fhir/packages/hl7.fhir.us.core#{old_version}/package') #latest version
# fhir_r4_pkg = Path(r'/Users/ehaas/.fhir/packages/hl7.fhir.r4.core#4.0.1/package')
#f_name = 'us-core-patient'
#n_path = '/Users/ehaas/.fhir/packages/hl7.fhir.us.core#dev' # package file keeps disappearing
#in_path = '//ERICS-AIR-2/ehaas/Documents/FHIR/US-Core-R4/output'  # use local build dir.
for i, sd in enumerate(new_pkg.glob('StructureDefinition-us-core-*.json')):
    sd_obj = loads(sd.read_text())  #get sd as fhir object
    new_ig_profiles.append((sd_obj['title'],sd_obj['url']))
print(len(new_ig_profiles))
# pprint(sorted(new_ig_profiles))
for i, sd in enumerate(old_pkg.glob('StructureDefinition-us-core-*.json')):
    sd_obj = loads(sd.read_text())  #get sd as fhir object
    old_ig_profiles.append((sd_obj['title'],sd_obj['url']))
print(len(old_ig_profiles))
# pprint(sorted(old_ig_profiles))
new_profiles = [i for i in new_ig_profiles if i not in old_ig_profiles]
print(len(new_profiles))
pprint(sorted(new_profiles))

deleted_profiles = [i for i in old_ig_profiles if i not in new_ig_profiles]
print(len(deleted_profiles))
pprint(sorted(deleted_profiles))
```
