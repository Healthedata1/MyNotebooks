Cycles through all markdown files in pagecontent, intro-notes and includes and creates a list of all words that match the regex pattern "\[.*?\]"

```python
import os
import re
from pathlib import Path

# Define the directory containing markdown files
directories = ["/Users/ehaas/Documents/FHIR/US-Core/input/pagecontent","/Users/ehaas/Documents/FHIR/US-Core/input/intro-notes","/Users/ehaas/Documents/FHIR/US-Core/input/includes","/Users/ehaas/Documents/FHIR/US-Core/input/data"]  # Update this path as needed

# list of file to igoore...
ignore_list = ["link-list.md","page-link-list.md","r4-link-list.md"]
# Regex pattern for words in square brackets
pattern = r"\[.*?\]"

# List to store all matches
bracketed_words = []
```

```python

# Iterate through all .md files in the directory
for directory in directories:
  for file_path in Path(directory).glob("*.md"):
    
    if file_path.name not in ignore_list:
      # print(file_path, type(file_path))    
      with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        # Find all matches in the file content
        matches = re.findall(pattern, content)
        bracketed_words.extend(matches)

# Remove duplicates and sort
unique_words = sorted(list(set(bracketed_words)))

# Print the results

print("Bracketed words found:")
print(len(unique_words))
for i in unique_words:
  print(i)

# Optionally, save the results to a ≠≠
# with open("bracketed_words.txt", "w", encoding="utf-8") as output_file:
#     for word in unique_words:
#         output_file.write(f"{word}\n")
```
