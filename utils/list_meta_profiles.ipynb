## List all meta profile for examples

- read file from folder
- convert to dict 
- find all meat.profile element
- create a dict file with example id as key and list of profiles it conforms to as the value.
- save fas yaml file back 

```python
from pathlib import Path
from json import dumps, loads
from yaml import load, dump, SafeLoader
from fhir.resources import construct_fhir_element

in_path = Path(r'/Users/ehaas/Documents/FHIR/davinci-alerts/input/examples-yaml')
out_path = Path(r'/Users/ehaas/Documents/FHIR/davinci-alerts/input/data/example_profile_list.yml')

file_comment = '''
# ================================================================================================
# This file is generated by the python script at 
# `https://github.com/Healthedata1/MyNotebooks/blob/master/utils/Add_meta.ipynblist_meta_profiles.ipynb`
# ================================================================================================
'''


def recursive_items(dictionary):
    for key, value in dictionary.items():
        if type(value) is dict:
            # print(f"key {key} value is dict")
            if value:  # Check if dictionary is not empty
                yield from recursive_items(value)
        elif type(value) is list:
            # print(f"key {key} value is list")
            for item in value:
                if type(item) is dict:
                    # print(f"key {item} value is dict")
                    yield from recursive_items(item)
                elif key == "profile":
                    yield (key, item)
        elif key == "profile":
            yield (key, item)

example_profile_list = {}
for f in Path(in_path).glob('*.yml'):
  ex_profiles = []
  print(f)
  ex_obj = load(f.read_text(),Loader=SafeLoader)
  ex_name = [i['valueString'] for i in ex_obj['meta']['extension'] if i['url'] =='http://hl7.org/fhir/StructureDefinition/instance-name'][0]
  print(ex_name)
  for key, value in recursive_items(ex_obj):
    # print(f"{key}: {value}")
    ex_profiles.append(value)
  example_profile_list[ex_name] = ex_profiles

example_profile_list = f'{file_comment}\n\n{dump(example_profile_list,indent=2)}'
out_path.write_text(example_profile_list)
```
