### Create a link list of All FHIR Resources
 - change path to change version
 - note that list will be same version as IG

```python
from pathlib import Path
from json import loads, dumps
my_path = Path(r'/Users/ehaas/.fhir/packages/hl7.fhir.r4.core#4.0.1/package') #Core FHIR
my_path
```

```python
for f in sorted(my_path.glob('Struct*.json')):
    # print(f)
    obj = loads(f.read_text())
    try:
      if obj['baseDefinition'] == 'http://hl7.org/fhir/StructureDefinition/DomainResource': print(f"[{obj['name']}]: {{{{site.data.fhir.path}}}}{obj['url'].split('http://hl7.org/fhir/StructureDefinition/')[-1].lower()}.html")
    except KeyError:
       pass
```
