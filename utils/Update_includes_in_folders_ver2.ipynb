### Replace Text from file

- get file(s) from path
- replace string in file
- save file (overwriting old)

```python
from pathlib import Path
import re
```

```python
# Define the regex pattern and replacement string
regex_pattern = r"(\{% include.*?)\.xhtml %\}"
regex_pattern_if_exist = r"(\{% include.*?)-en\.xhtml %\}"
replacement_string = r"\1-en.xhtml %}"

path = Path(r'/Users/ehaas/Documents/FHIR/US-Core/input/pagecontent')
path
```

```python
for f in path.glob('*.md'):
  in_text = f.read_text()
  try:
      # Read the file content
      content = f.read_text(encoding="utf-8")
      # Check if the replacement string (e.g., list-*-en.xhtml) already exists
      if re.search(regex_pattern_if_exist, content):
          print(f"No changes in {f}: replacement pattern '{regex_pattern_if_exist}' already exists")
          continue
      # Perform the regex-based string replacement
      new_content, count = re.subn(regex_pattern, replacement_string, content)
      # Write the updated content back to the file if changes were made
      if count > 0:
          f.write_text(new_content, encoding="utf-8")
          print(path.name)
          print (f'IG file: {f}')
          print(f"Updated {f}: replaced {count} occurrence(s) matching '{regex_pattern}' with '{replacement_string}'")
          print("----- New Content Start -----")
          print(new_content)
          print("----- New Content End -----")
      else:
          pass
          print(f"No changes in {f}: patterns not found")
  except (IOError, UnicodeDecodeError) as e:
      print(f"Error processing {f}: {e}")
```
