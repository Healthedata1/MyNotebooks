## Create Markdown Text for SearchParameters Page

- Using Jinja2 Template create markdown file for searchparameters page
- Use spreadsheet as source 
- Omits custom SP expecting to be listed separately

```python
from pandas import *
from pathlib import Path
from openpyxl import load_workbook
from jinja2 import Environment, FileSystemLoader, select_autoescape
from IPython import display
```

```python
#ig_source_path = "//ERICS-AIR-2/ehaas/Documents/FHIR/US-Core-R4/input/"  # for pc
in_path = Path(r"/Users/ehaas/Documents/FHIR/US-Core/input/resources_spreadsheets/")    # for mac
out_path = Path(r"/Users/ehaas/Documents/FHIR/US-Core/input/includes/uscore_sp_list.md")    # for mac
#out_path = Path() / '_includes' / 'uscore_sp_list.md' # temp path for testing
in_file ="uscore-server.xlsx"
ig_base = "us-core"

skip_types = ['Questionnaire',]
# custom_sp = ['race','ethnicity','gender-identity',]
```

### Get spreadsheet data using Pandas

```python
xls = ExcelFile(in_path / in_file )
df = read_excel(xls,'sps',na_filter = False)
df = df.sort_values(by=['base','code'])
data = [i for i in df.itertuples(index=True)
                  if '!' not in i.base]
```

```python
r_list = {r.base for r in data}
my_string = ''''''

def append_to_string(my_string,s):
    my_string = my_string + '\n' + s
    return
for r in sorted(r_list):
    s = f'##### {r}\n'
    my_string = my_string + s
    for d in data:
        if d.base == r and d.exists == 'Y':  # d.exists == N for custom SP listed separately in US Core
            new_content_tag = '{:.new-content}' if d.is_new else ''
            s = f'- { new_content_tag}[{d.code}](SearchParameter-{ig_base}-{r.lower()}-{d.code.lower().replace("_","")}.html)\n'
            my_string = my_string + s
    my_string = my_string + '\n'
# add comment to line 1
my_string = f'<!-- Source = {in_path}. This file is generated by SearchParameterListMaker.ipynb. Do not edit directly. -->\n\n{my_string}'
print(my_string)
display.display_markdown(my_string, raw=True)
```

## write to file

```python
out_path.write_text(my_string)
```
